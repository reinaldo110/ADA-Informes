# 1. NP-Hard

## 1.1 ¿Qué quiere decir que un problema es NP-Hard? 

Para un mejor entendimiento, primero se explicará cada grupo de problemas existente, para finalmente, explicar que es un problema NP-Hard.

En primer lugar, un algoritmo resuelve un problema de manera eficiente si el coste de su tiempo es un polinomio que está en la entrada del problema, osea el tamaño de esta. Este tipo de problemas entran en la categoría de problemas P (problemas que se resuelven en tiempo polinomial). 

En segundo lugar, existen problemas que no se pueden resolver en un tiempo polinomial, pero aún así pueden verificar su solución en un tiempo polinómico, estos problemas se conocen como problemas NP.

En tercer lugar, los problemas NP-Completos son aquellos problemas más difíciles del conjunto NP.

Finalmente, un problema NP-Hard es un problema de decisión que contiene problemas H, tales que cualquier problema L perteneciente a NP puede ser reducido en tiempo polinomial a H. Además, estos problemas tienen una complejidad comparable a los problemas NP-Completos, pero no necesariamente pertenecen al conjunto NP.

## 1.2 ¿Qué quiere decir que $P ≠ NP$?

Un término que se ve a menudo es $P ≠ NP$, ¿pero qué significa?

$P ≠ NP$ es una conjetura, esto significa que se cree que es correcto, pero no se aprobado, aún. En pocas palabras, esta conjetura significa que no es posible resolver problemas pertenecientes al conjunto NP en tiempo polinomial. Por eso se dice que el conjunto P es distinto al NP. Sin embargo, si alguna vez se logra resolver, aunque sea sólo un problema NP en tiempo polinomial, entonces se probaría que la conjetura es falsa y que en verdad todos los problemas NP pertenecerían al conjunto P.

Debido a todo lo anterior, es que se dice que $P ≠ NP$ es una impresición, pero una que se acepta, esto debido a que no se ha podido probar con exactitud que esta es verdadera.

# 2. Problemas NP-Hard

## 2.1 *Knapsack*

**Entrada:** Conjunto de objetos, cada uno con un valor y peso específico de las formas $V = [v_1, v_2, ..., v_n]$ y $P = [p_1, p_2, ..., p_n]$ respectivamente. También se recibe un peso máximo específico de la mochila $W$ y la cantidad de objetos presentes en el conjunto.

**Salida:** Valor máximo que se puede almacenar en la mochila ingresada, de forma que este valor corresponda a la solución óptima del problema. 

El objetivo de este problema es el de calcular la máxima ganancia que se puede obtener al ingresar una cantidad limitada de objetos en una mochila con un almacenamiento limitado.

## 2.2 *Set Cover*

**Entrada:** Conjunto $S = (a_1, a_2, a_3, ..., a_n)$, conjunto de sub-conjuntos de $S$: $[B_1, B_2, ..., B_q]$ y un entero positivo $k$.

**Salida:** Sub-conjuntos que permiten la cantidad mínima de sub-conjuntos para generar $S$.

El objetivo del problema de cobertura de conjuntos es encontrar todos los conjuntos cuya unión contiene a todos los elementos de $S$.

# 3. Demostración de que los problemas anteriores son NP-Hard (usando reducción)

## 3.1 Cómo y por qué el problema *Subset Sum* se puede reducir al problema de la mochila (*Knapsack*)

### *Subset Sum*

**Entrada:** Conjunto de números enteros no negativos $[a_1,a_2,...,a_n]$ y un número entero objetivo $s$.

**Salida:** Una respuesta *booleana* (SI / NO), dependiendo de si hay o no hay una secuencia de números que sumados sean igual a $s$.

El problema de la Suma de Sub-conjuntos (*Subset Sum*) consiste en que dado un conjunto de números enteros y un valor objetivo, se determine si en el conjunto de entrada se encuentra una secuencia, que al ser sumada, sea igual al valor objetivo.

Cómo se mencionó antes, el problema de suma de sub-conjuntos tiene como entrada a un conjunto de números enteros y un número $s$. Esto puede ser utilizado para reducir este problema al problema de la mochila.

**¿Por qué?**

Esto se puede hacer debido a que el resultado obtenido al realizar el problema de la mochila, puede ser utilizado para resolver el problema de suma de sub-conjuntos, ya que la salida del algoritmo *Knapsack*, equivale al mayor número de objetos que se pueden almacenar en una mochila, maximizando el valor almacenado, sin sobrepasar el peso permitido. 

**¿Cómo?**

Para lograr la reducción del problema, se deberían seguir los siguientes pasos.

1. Ingresar los datos de entrada del problema de suma de sub-conjuntos.

2. Estos datos también se utilizan como datos de entrada para el problema de la mochila.

3. Se resuelve el problema de la mochila y se retornan los datos de salida, los cuáles sirven para dar solución al problema de suma de sub-conjuntos.

4. Con estos resultados, se puede resolver el problema de suma de sub-conjuntos y así retornar su solución.

A continuación se muestra un ejemplo para mejor comprensión.

1. Se ingresa el conjunto $[3, 4, 5]$ y un número $s$ igual a $9$.

2. Estos datos se ingresan al problema de la mochila como si fueran los pesos de los objetos y el peso máximo que soporta la mochila, respectivamente. Además, se hace que los pesos de los objetos sean iguales a sus valores.

  val = [3, 4, 5]

  wt = [3, 4, 5]

  W = 9

3. La mochila resuelve su sub-problema, retornando el valor $9$, que representa la máxima ganancia que se puede obtener al guardar los objetos en la mochila sin sobrepasar el peso que esta soporta.

4. Con este resultado, se puede resolver el problema de la suma de sub-conjuntos del inicio, ya que el resultado arrojado por el problema de la mochila se puede comparar con el valor objetivo ingresado $s$ para saber conocer la solución del problema.

5. Finalmente, el resultado arrojado por la suma de sub-conjuntos debería ser que sí existen números en el conjunto $[3, 4, 5]$ tal que su suma sea igual a $9$, ya que la respuesta dada por el problema de la mochila coincide con el valor objetivo del principio.

**NP-Hard**

Cómo el problema de suma de sub-conjuntos, que sabemos es NP-Hard, es reducible al problema de la mochila, entonces, por la conjetura, se puede concluir que el problema de la mochila también es NP-Hard.

## 3.2 Algoritmo que utilice reducción para resolver *Subset Sum* usando *Knapsack*

Para realizar la reducción se deben realizar los siguientes pasos:

**Input (*Subset Sum*)**

1. El algoritmo *Subset Sum* recibe como entrada un conjunto de enteros y un valor objetivo.

**Preproceso**

2. Lo primero que haría este algoritmo sería tomar estos datos de entrada y transformarlos en los datos de entrada del problema de la mochila. Para esto, generaría dos arreglos que representen el peso y el valor de los objetos a guardar en la mochila y los llenaría con los datos del conjunto recibido, ambos arreglos deben ser iguales.

3. También, el valor objetivo o suma a buscar de *Subset Sum*, se transformaría en la capacidad máxima de la mochila.

**Sub-rutina para *Knapsack***

4. Después se llama a la función *Knapsack* y le entregamos los datos generados en el preproceso.

5. Esta función nos va a retornar un valor, el cuál va a ser el valor máximo posible que respete la capacidad máxima de la mochila maximimizando la ganancia.

**Postproceso**

6. Ahora la función *Subset Sum* recibe la solución retornada por el problema de la mochila (que corresponde a la ganancia máxima) y la transforma en su propia solución (una suma que sea igual al valor objetivo de la entrada).

**Output (*Subset Sum*)**

7. Si la solución entregada por *Knapsack* es igual al valor objetivo, el problema original del *Subset Sum* habrá sido resuelto correctamente y por ende retorná un SI, en caso contrario, retornará un NO.

## 3.3 Cómo y por qué el problema *Vertex Cover* se puede reducir al problema *Set Cover*

## *Vertex Cover*

**Entrada:** Grafo no dirigido $G(V, E)$.

**Salida:** Una respuesta *booleana* (SI / NO), de si existe o no una cobertura de vértices $V´$ para el grafo $G(V, E)$.

El problema de la Cobertura de Vértices (*Vertex Cover*) consiste en que dado un grafo no dirigido, se determine si hay un conjunto de vértices que permita conectar cada vértice de este conjunto con cada otro vértice del grafo.

El problema de *Vertex Cover* se puede reducir al problema de *Set Cover* modificando la entrada que este recibe.

**¿Por qué?**

Dado una entrada de *Vertex Cover*: Un grafo no dirigido $G = (V, E)$ y un entero positivo $k$.

*Set Cover* tiene la siguiente entrada: Un conjunto $S = (a_1, a_2, a_3, ..., a_n)$, un conjunto de sub-conjuntos de $S$: $[B_1, B_2, ..., B_q]$ y un entero positivo $k$.

Nosotros sabemos que el problema de *Vertex Cover* debe cubrir todos los arcos del grafo. En el problema de *Set Cover*, nosotros necesitamos cubrir todos los elementos del conjunto $S$, entonces vamos a convertir cada arco de $G$ en un elemento de un conjunto de $S$ e intentar cubrir el conjunto $S$.

En el problema de *Vertex Cover* nosotros necesitamos cubrir todos los arcos utilizando un numero limitado de nodos. En el problema de *Set Cover* necesitamos cubrir los elementos de un conjunto utilizando un numero limitado de sub-conjuntos. Entonces esto nos da la idea de "mapear" los vértices del problema de *Vertex Cover* en sub-conjuntos del problema de *Set Cover*.

**¿Cómo?**

Cada arco $E_i$ va a corresponder a un elemento $a_i$ del conjunto $S$.

Cada vértice $V_i$ va a corresponder un sub-conjunto de $S$, llamado $B_i$.

El conjunto $B_i$ va a contener todos los elementos $a_j$, los cuáles corresponden a los arcos $e_j$, los cuáles son los que están unidos al vértice $V_i$.

Entonces, aplicando la solución propuesta por *Set Cover*, nos va a retornar los sub-conjuntos que cubren todo $S$. Estos sub-conjuntos van a ser equivalentes a los vértices que cubren todos los arcos en el grafo del problema *Vertex Cover*.

**NP-Hard**

Cómo el problema de cobertura de vértices, que sabemos es NP-Hard, es reducible al problema del conjunto de cobertura, entonces, por la conjetura, se puede concluir que el problema del conjunto de cobertura también es NP-Hard.

## 3.4 Algoritmo que utilice reducción para resolver *Vertex Cover* usando *Set Cover*

Para realizar esta reducción se deben realizar los siguientes pasos:

**Input (*Vertex Cover*)**

1. El algoritmo recibe como entrada un grafo no dirigido y un entero positivo.

**Preproceso** 

2. Este algoritmo convierte los vértices del grafo recibido en sub-conjuntos de $S$. Primero se considera que todos los arcos del grafo sean el conjunto $S$. Luego los vértices y sus arcos adyacentes (todo arco que esta conectado al vértice) van a representar sub-conjuntos de $S$.

3. Los nuevos sub-conjuntos se conoceran como $B_i$.

**Sub-rutina para *Set Cover***

4. Llamamos el algoritmo *Set Cover* con los datos del proceso anterior como su entrada.

5. Esta función nos va a retornar los sub-conjuntos que permiten la cantidad mínima de sub-conjuntos para generar $S$.

**Postproceso**

6. La función *Vertex Cover* recibe la solución retornada por *Set Cover* y la analiza para ver que solución retornará.

**Output (*Vertex Cover*)**

7. Debemos recordar que definimos a los sub-conjuntos como vértices, entonces se debe realizar una comparación de los sub-conjuntos que tuvimos como salida de *Set Cover* con los vértices de la salida de *Vertex Cover*, si son los mismos vértices, retornará un SI, en caso contrario, un NO.

# 4. Algoritmos sencillos para resolver los problemas utilizados para reducir los problemas NP-Hard escogidos

## 4.1 Algoritmo para resolver el problema *Knapsack* y su tiempo de ejecución

### Algoritmo

El problema de la mochila puede ser resuelto de forma sencilla de la siguiente manera, de forma recursiva.

1. Como entrada vamos a tener la capacidad máxima de la mochila **W**, el arreglo con el peso de los objetos **wt**, el arreglo con el valor de los objetos **val** y la cantidad total de objetos $n$.

2. Como caso base vamos a tener que si **W = 0** o **n = 0**, se debe retornar 0.

3. Si el peso del n-ésimo objeto hace que la suma total de pesos sea mayor que la capacidad máxima de la mochila, no se puede incluir este objeto, vamos a llamar *Knapsack* recursivamente con $n-1$ objetos.

4. En caso contrario, se debe retornar el máximo valor entre los siguientes casos:

  1. El n-ésimo objeto es incluido en la mochila.

  2. El n-ésimo objeto no es incluido en la mochila.

5. Para comprobar esto se utilizará la función *max* entre:

  1. La suma entre el valor del objeto que se incluye y la función *Knapsack* con una capacidad máxima menos el peso del objeto y $n - 1$ objetos.

  2. La función *Knapsack* con $n - 1$ objetos.

### Tiempo de ejecución

El tiempo de ejecución del algoritmo planteado para resolver este problema, sería de $O(2^n)$ debido a la gran cantidad de sub-problemas que se deberían de resolver. Lo anterior debido a que es una solución recursiva para el problema.

## 4.2 Algoritmo para resolver el problema *Set Cover* y su tiempo de ejecución

### Algoritmo

Se pueden hacer los siguientes pasos para resolver el problema del *Set Cover*.

Dada la siguiente entrada: Un conjunto $S = (a_1, a_2, a_3, ..., a_n)$, un conjunto de sub-conjuntos de $S$: $[B_1, B_2, ..., B_q]$ y un entero positivo $k$.

1. Se crea un arreglo $I$ inicializado vacío.

2. Después se van recorriendo los sub-conjuntos mientras $I$ sea menor que $S$, viendo cuáles son aptos para cubrir todo el conjunto $S$. Se deben recorrer todos los sub-conjuntos para saber cuantos sub-conjuntos usaríamos para cubrir $S$, es decir, debemos asegurarnos de elegir la cantidad mínima de sub-conjuntos y que estos cubran a $S$.

3. Los sub-conjuntos aptos serán guardados en $I$ y luego este arreglo será retornado.

### Tiempo de ejecución

El tiempo de ejecución de este algoritmo planteado sería el tamaño del conjunto $S$, osea, $O(n)$, debido a que se debe ir recorriendo en un ciclo iterativo todos los sub-conjuntos hasta que $I$ sea del mismo tamaño que $S$.
